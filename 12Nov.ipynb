```python
s='python pytohn pyton pytno pytkl'
l=s.split()
print(l)
l=sorted(l)
print(l)
if len(l[0])>len(l[-1]):
    length=len(l[-1])
else:
    length=len(l[0])
ans=""
for i in range(length):
    if l[0][i]==l[-1][i]:
        ans+=l[0][i]
    else:
        break
print(ans)
```

```python
def f(s):
    stack=[]
    d={')':'(',']':'[','}':'{'}
    if len(s)%2!=0:
        return False
    for i in s:
        if i in '({[':
            stack.append(i)
        else:
            if len(stack)==0:
                return False
            elif d[i]!=stack.pop():
                return False
    return len(stack)==0
s=input("Enter: ")
if f(s):
    print('valid')
else:
    print('invalid')
```
